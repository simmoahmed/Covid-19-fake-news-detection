```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import f1_score
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
```

```python
train = pd.read_csv('data/postproc_train.csv')
val = pd.read_csv('data/postproc_val.csv')
test = pd.read_csv('data/postproc_test.csv')
```

```python
train_X = train.iloc[:, :-1].values
train_y = train.iloc[:, -1].values
val_X = val.iloc[:, :-1].values
val_y = val.iloc[:, -1].values
test_X = test.iloc[:, :-1].values
test_y = test.iloc[:, -1].values
```

```python
cv = ShuffleSplit(n_splits=10, test_size=0.3, random_state=0)

model = LogisticRegression(solver='liblinear', max_iter=400, penalty='l2', random_state=0)
# scores = cross_val_score(model, train_X, train_y, cv=cv, scoring='f1_micro')
# scores.mean()
```

```python
model.fit(train_X, train_y)
```

```python
# val_scores = cross_val_score(model, val_X, val_y, cv=cv, scoring='f1_micro')
# val_scores.mean()
```

```python
val_predictions = model.predict(val_X)
```

```python
f1_score(val_y, val_predictions, average='micro')
```
