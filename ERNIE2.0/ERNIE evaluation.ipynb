```
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, confusion_matrix
```

```
val_df = pd.read_csv("/content/drive/MyDrive/Constraint/data/Constraint_Val.csv")
# val_df = val_df[["id"]]
val_df.head()
```

```
test_df = pd.read_csv("/content/drive/MyDrive/Constraint/data/english_test_with_labels.csv")
# test_df = test_df[["id"]]
test_df.head()
```

```
ernie_val_df = pd.read_csv("/content/drive/MyDrive/Constraint/Boosting/Intermediate/ERNIE/val_ernie2.0_results.csv")
ernie_val_df = ernie_val_df.drop(['Unnamed: 0', 'Label'], 1)
ernie_val_df.head()
```

```
ernie_test_df = pd.read_csv("/content/drive/MyDrive/Constraint/Boosting/Intermediate/ERNIE/test_ernie2.0_results.csv")
ernie_test_df = ernie_test_df.drop('Unnamed: 0', 1)
ernie_test_df.head()
```

```
def calc_argmax(row):
  val = np.argmax([[row["Model4_class0"], row["Model4_class1"]]], 1)
  return 'real' if val==0 else 'fake'

ernie_val_df["predicted_label"] = ernie_val_df.apply(lambda x: calc_argmax(x), 1)
ernie_test_df["predicted_label"] = ernie_test_df.apply(lambda x: calc_argmax(x), 1)
```

```
final_val_df = pd.concat([val_df, ernie_val_df[["predicted_label"]]], 1)
final_val_df.head()
```

```
final_test_df = pd.concat([test_df, ernie_test_df[["predicted_label"]]], 1)
final_test_df.head()
```

```
accuracy_score(final_val_df['label'].values, final_val_df['predicted_label'].values)
```

```
precision_score(final_val_df['label'].values, final_val_df['predicted_label'].values, average='micro')
```

```
recall_score(final_val_df['label'].values, final_val_df['predicted_label'].values, average='micro')
```

```
f1_score(final_val_df['label'].values, final_val_df['predicted_label'].values, average='micro')
```

```
accuracy_score(final_test_df['label'].values, final_test_df['predicted_label'].values)
```

```
precision_score(final_test_df['label'].values, final_test_df['predicted_label'].values, average='micro')
```

```
recall_score(final_test_df['label'].values, final_test_df['predicted_label'].values, average='micro')
```

```
f1_score(final_test_df['label'].values, final_test_df['predicted_label'].values, average='micro')
```
