```
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score
```

```
test_df = pd.read_csv("/content/drive/MyDrive/Constraint/data/english_test_with_labels.csv")
# test_df = test_df[["id"]]
test_df
```

```
ernie_df = pd.read_csv("/content/drive/MyDrive/Constraint/Boosting/Intermediate/ERNIE/test_ernie2.0_results.csv")
ernie_df.head()
```

```
def calc_argmax(row):
  val = np.argmax([[row["Model4_class0"], row["Model4_class1"]]], 1)
  return 'real' if val==0 else 'fake'

ernie_df["predicted_label"] = ernie_df.apply(lambda x: calc_argmax(x), 1)
```

```
ernie_df = pd.concat([test_df[["id", "label"]], ernie_df[["predicted_label"]]], 1)
ernie_df.head()
```

```
# ernie_df.to_csv('/content/drive/MyDrive/Constraint/Submission_v5.csv', index=False)
```

```
f1_score(ernie_df['label'].values, ernie_df['predicted_label'].values, average='micro')
```
